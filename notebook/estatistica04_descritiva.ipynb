```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ["[2,6)","[6,10)","[10,14)","[14,18)"]
y = [30, 35, 25, 10]
```

```python
fig, ax = plt.subplots(figsize=(12,6))

plt.bar(x,y, width = 1.0, color=['green', 'blue', 'yellow', 'red'])

ax.set_xlabel('intervalos')
ax.set_ylabel('porcentagens')
ax.bar_label(ax.containers[0])

fig.savefig('images/graf_hist_freq_salarios.jpg')

plt.show()
```
